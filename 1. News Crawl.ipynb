```python
# Create driver
driver = webdriver.Chrome()
# Get a page
driver.get('https://cafef.vn/thi-truong-chung-khoan.chn')
# Feed the source to BeautifulSoup
soup = BeautifulSoup(driver.page_source)
l = soup.find_all('div', attrs={'class':'tlitem'})
```

```python
start_time = time.time()
for i in range(1, 800):
    if i % 50 == 0:
        end_time = time.time()
        print(f"Crawl item from {i-50+1} to {i} in {end_time-start_time} seconds")
        time.sleep(10)
        start_time = time.time()       
    driver.execute_script("document.getElementsByClassName('btn-viewmore')[0].click()") # Click view more button
l = BeautifulSoup(driver.page_source).find_all('div', attrs={'class':'tlitem'})
```

```python
res = []
tim_mod=[]

for i, article in enumerate(l):   
    title = article.find('h3').find('a').text
    tim = article.find('span', class_= 'time-ago')['title']
    link = article.find('h3').find('a')['href']
    res.append([tim, title, link])
```

```python
pd.DataFrame(res, columns=['Time','Headline','Link']).to_excel('data.xlsx')
```

```python
df['content']=""
for index, row in df.iterrows():
    url = row['Link']
    # Gửi yêu cầu GET đến URL và lấy nội dung của trang
    response = requests.get(url)
    content = response.content

    # Sử dụng BeautifulSoup để phân tích và tìm kiếm trong nội dung HTML
    soup = BeautifulSoup(content, "html.parser")

    # Tìm phần tử có class "detail-content afcbc-body"
    detail_content = soup.find("div", class_="contentdetail")
    
    if detail_content != None:
        df.at[index, 'content'] = detail_content.get_text()
    else:
        df.at[index, 'content'] = "Không có thông tin"
    time.sleep(5)
```
